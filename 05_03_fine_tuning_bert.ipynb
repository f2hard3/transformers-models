```python
custom_module = 'savasy/bert-base-turkish-ner-cased'
ner = pipeline('ner', model=custom_module, tokenizer=custom_module) # named entity recognition
ner
```

```python
bert_qa = BertForQuestionAnswering.from_pretrained('bert-base-uncased')
bert_qa
```

```python
model_name = 'deepset/roberta-base-squad2'
qa = pipeline(model=model_name, tokenizer=model_name, revision='v1.0', task='question-answering')
qa
```

```python
qa.model
```

```python
sequence = 'Where is Sunggon living these days?', 'Sunggon lives in Tokyo, Japan but Changmin lives in Seoul, Korea'
qa(*sequence)
```

```python
sequence = 'Where is Changmin living these days?', 'Sunggon lives in Saitama, Japan but Changmin lives in Seoul, Korea'
qa(*sequence)
```
